# Needed Action

   # 1- Remove Outliers
   # 2- Replace with something else(q1/q3)
   

```python
df.describe()
```

```python
df['DiscountPrice (in Rs)'].fillna(824 , inplace = True)
```

```python
q1 = df.quantile( q = .25)
q3 = df.quantile( q = .75)
UW = q3 + (1.5*(q3-q1))
LW = q1 - (1.5*(q3-q1))
```

```python
for i in range(len(df['DiscountPrice (in Rs)'])):
    if df['DiscountPrice (in Rs)'][i] >= UW['DiscountPrice (in Rs)']:
        df['DiscountPrice (in Rs)'][i] = UW['DiscountPrice (in Rs)']
    elif df['DiscountPrice (in Rs)'][i] <= LW['DiscountPrice (in Rs)']:
        df['DiscountPrice (in Rs)'][i] = LW['DiscountPrice (in Rs)']
```

```python
sns.boxplot(data = df, y = 'DiscountPrice (in Rs)')   # after removing outliers
```
